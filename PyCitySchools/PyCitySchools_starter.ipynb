```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
```

## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average math score 

* Calculate the average reading score

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Calculate the percentage of students who passed math **and** reading (% Overall Passing)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
# ln - Getting values for:
#      - total schools
#      - total students
#      - total budget
#      - average math score
#      - average reading score
#      - % passing math
#      - % passing reading
#      - % overall passing

total_school = len(school_data_complete["school_name"].unique())

total_students = school_data_complete["student_name"].count()
#mapped_total_students = total_students.map("{:,}".format)

total_budget = school_data_complete["budget"].unique().sum()
#mapped_total_budget = total_budget.map("{:,.2f}".format)

avg_math = school_data_complete["math_score"].mean()

avg_read = school_data_complete["reading_score"].mean()

passing_math = school_data_complete.loc[school_data_complete["math_score"] >= 70, :]
stu_pass_math = (len(passing_math)/total_students) * 100

passing_read = school_data_complete.loc[school_data_complete["reading_score"] >= 70, :]
stu_pass_read = (len(passing_read)/total_students) * 100

overall_pass = school_data_complete.loc[((school_data_complete["math_score"] >= 70) & (school_data_complete["reading_score"] >= 70)), :]
stu_overall_pass = (len(overall_pass)/total_students) * 100

# Creating a dataframe of values of all necessary categories
district_summary_df = pd.DataFrame({"Total Schools": [total_school],
                            "Total Students": [total_students],
                            "Total Budget": [total_budget],
                            "Average Math Score": [avg_math],
                            "Average Reading Score": [avg_read],
                            "% Passing Math": [stu_pass_math],
                            "% Passing Reading": [stu_pass_read],
                            "% Overall Passing": [stu_overall_pass]})


mapped_district_summary_df = district_summary_df.copy()
mapped_district_summary_df["Total Students"] = mapped_district_summary_df["Total Students"].map("{:,}".format)
mapped_district_summary_df["Total Budget"] = mapped_district_summary_df["Total Budget"].map("${:,.2f}".format)
mapped_district_summary_df
```

## School Summary

* Create an overview table that summarizes key metrics about each school, including:
  * School Name
  * School Type
  * Total Students
  * Total School Budget
  * Per Student Budget
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * % Overall Passing (The percentage of students that passed math **and** reading.)
  
* Create a dataframe to hold the above results

```python
# ln - Getting the grouped dataframe of:
#      - school type of each school
#      - total students of each school
#      - total school budget of each school
#      - per student budget of each school
#      - average math score of each school
#      - average reading score of each school
#      - % passing math of each school
#      - % passing reading of each school
#      - % overall passing of each school

school_copy_summary = school_data_complete.copy()

school_name = school_copy_summary.set_index("school_name").groupby("school_name")

school_type = school_copy_summary.groupby("school_name")["type"].first()

student_count = school_name["Student ID"].count().astype("int64")

school_budget = school_copy_summary.groupby("school_name")["budget"].unique().astype("float")

per_student_budget = school_budget / student_count

avg_math_score = school_copy_summary.groupby("school_name")["math_score"].mean().astype("float")

avg_read_score = school_copy_summary.groupby("school_name")["reading_score"].mean().astype("float")

percent_passing_math = (school_copy_summary.loc[(school_copy_summary["math_score"] >= 70), :].groupby("school_name")["Student ID"].count() / student_count) * 100

percent_passing_read = (school_copy_summary.loc[(school_copy_summary["reading_score"] >= 70), :].groupby("school_name")["Student ID"].count() / student_count) * 100

percent_passing_overall = school_copy_summary.loc[((school_copy_summary["math_score"] >= 70) & (school_copy_summary["reading_score"] >= 70)), :]
percent_passing_overall = (percent_passing_overall.groupby("school_name")["Student ID"].count() / student_count) * 100

# Creating a dataframe of dataframes of all necessary categories
school_summary = pd.DataFrame({"School Type": school_type,
                               "Total Students": student_count,
                               "Total School Budget": school_budget,
                               "Per Student Budget": per_student_budget,
                               "Average Math Score": avg_math_score,
                               "Average Reading Score": avg_read_score,
                               "% Passing Math": percent_passing_math,
                               "% Passing Reading": percent_passing_read,
                               "% Overall Passing": percent_passing_overall})
school_summary

mapped_school_summary = school_summary.copy()
mapped_school_summary["Total School Budget"] = mapped_school_summary["Total School Budget"].map("${:,.2f}".format)
mapped_school_summary["Per Student Budget"] = mapped_school_summary["Per Student Budget"].map("${:.2f}".format)
mapped_school_summary
```

## Top Performing Schools (By % Overall Passing)

* Sort and display the top five performing schools by % overall passing.

```python
# ln - Sort values of dataframe based on % overall passing category for top 5 schools
top_schools_summary = school_summary.sort_values("% Overall Passing", ascending = False).head()
top_schools_summary

mapped_top_schools_summary = top_schools_summary.copy()
mapped_top_schools_summary["Total School Budget"] = mapped_top_schools_summary["Total School Budget"].map("${:,.2f}".format)
mapped_top_schools_summary["Per Student Budget"] = mapped_top_schools_summary["Per Student Budget"].map("${:.2f}".format)
mapped_top_schools_summary
```

## Bottom Performing Schools (By % Overall Passing)

* Sort and display the five worst-performing schools by % overall passing.

```python
# ln - Sort values of dataframe based on % overall passing category for bottom 5 schools
bottom_schools_summary = school_summary.sort_values("% Overall Passing", ascending = True).head()
bottom_schools_summary

mapped_bottom_schools_summary = bottom_schools_summary.copy()
mapped_bottom_schools_summary["Total School Budget"] = mapped_bottom_schools_summary["Total School Budget"].map("${:,.2f}".format)
mapped_bottom_schools_summary["Per Student Budget"] = mapped_bottom_schools_summary["Per Student Budget"].map("${:.2f}".format)
mapped_bottom_schools_summary
```

## Scores by School Spending

* Create a table that breaks down school performances based on average Spending Ranges (Per Student). Use 4 reasonable bins to group school spending. Include in the table each of the following:
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)

```python
# ln - Binning based on school spending per school
school_spending_summary = school_summary.copy()
spending_bin = [0, 585, 630, 645, 680]
budget_group_labels = ["<$585", "$585-630", "$630-645", "$645-680"]

spending_range_per_student = pd.cut(school_spending_summary["Per Student Budget"], bins = spending_bin, labels = budget_group_labels)
school_spending_summary["Spending Ranges (Per Student)"] = spending_range_per_student

spending_range_group = school_spending_summary.groupby("Spending Ranges (Per Student)")
spending_summary = spending_range_group[["Average Math Score", "Average Reading Score", "% Passing Math", "% Passing Reading", "% Overall Passing"]].mean()
spending_summary
```

## Scores by School Size

* Perform the same operations as above, based on school size.

```python
# ln - Binning based on school size per school
school_size_summary = school_summary.copy()
school_size_bin = [0, 1000, 2000, 5000]
size_group_labels = ["Small (<1000)", "Medium (1000-2000)", "Large (2000-5000)"]

school_size = pd.cut(school_size_summary["Total Students"], bins = school_size_bin, labels = size_group_labels)
school_size_summary["School Size"] = school_size

school_size_group = school_size_summary.groupby("School Size")
size_summary = school_size_group[["Average Math Score", "Average Reading Score", "% Passing Math", "% Passing Reading", "% Overall Passing"]].mean()
size_summary
```
