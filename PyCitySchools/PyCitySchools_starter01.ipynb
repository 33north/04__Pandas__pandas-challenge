```python
# Dependencies and Setup
import pandas as pd

# Filepath
schools_complete_filepath = "resources/schools_complete.csv"
students_complete_filepath = "resources/students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames
schools_complete_df = pd.read_csv(schools_complete_filepath)
students_complete_df = pd.read_csv(students_complete_filepath)
```

```python
# Combine the data into a single dataset.  
student_school_df = pd.merge(left = students_complete_df, right = schools_complete_df, how = "left", on = "school_name")
student_school_df
```

```python
student_school_df.columns
```

```python
student_school_df = student_school_df.rename(columns = {"Student ID": "student_id",
                                                        "School ID": "school_id"})
student_school_df.columns
```

## District Summary
* Calculate the total number of schools
* Calculate the total number of students
* Calculate the total budget
* Calculate the average math score 
* Calculate the average reading score
* Calculate the percentage of students with a passing math score (70 or greater)
* Calculate the percentage of students with a passing reading score (70 or greater)
* Calculate the percentage of students who passed math **and** reading (% Overall Passing)
* Create a dataframe to hold the above results
* Optional: give the displayed data cleaner formatting

```python
total_schools = len(student_school_df["school_name"].unique())
total_students = len(student_school_df["student_name"])
total_budget = student_school_df["budget"].unique().sum()
avg_math_score = student_school_df["math_score"].mean()
avg_read_score = student_school_df["reading_score"].mean()
pct_pass_math = ((len(student_school_df.loc[student_school_df["math_score"] >= 70, :]) / total_students) * 100)
pct_pass_read = ((len(student_school_df.loc[(student_school_df["reading_score"] >= 70), :]) / total_students) * 100)
overall_pass_math_read = ((len(student_school_df.loc[((student_school_df["math_score"] >= 70) & (student_school_df["reading_score"] >= 70)), :]) / total_students) * 100)

overview_df = pd.DataFrame({"Total Schools": [total_schools],
                            "Total Students": [total_students],
                            "Total Budget": [total_budget],
                            "Average Math Score": [avg_math_score],
                            "Average Reading Score": [avg_read_score],
                            "% Passing Math": [pct_pass_math],
                            "% Passing Reading": [pct_pass_read],
                            "% Overall Passing": [overall_pass_math_read]})
overview_df

# overview_dict = {"Total Schools": [total_schools],
#                  "Total Students": [total_students],
#                  "Total Budget": [total_budget],
#                  "Average Math Score": [avg_math_score],
#                  "Average Reading Score": [avg_read_score],
#                  "% Passing Math": [pct_pass_math],
#                  "% Passing Reading": [pct_pass_read],
#                  "% Overall Passing": [overall_pass_math_read]}
# overview_df = pd.DataFrame(overview_dict)
# overview_df
```

## School Summary
* Create an overview table that summarizes key metrics about each school, including:
  * School Name
  * School Type
  * Total Students
  * Total School Budget
  * Per Student Budget
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * % Overall Passing (The percentage of students that passed math **and** reading.)
* Create a dataframe to hold the above results

```python
stu_sch_df_copy = student_school_df.copy()
# print(stu_sch_df_copy)

school_count = len(stu_sch_df_copy["school_name"].unique())
# print(school_count)

school_name = stu_sch_df_copy["school_name"].unique()
school_name

school_name_group = stu_sch_df_copy.set_index("school_name").groupby(["school_name"])
school_type = school_name_group["type"].first()
student_count = school_name_group["student_id"].count().astype("int64")
school_budget = school_name_group["budget"].first()
per_stu_budget = school_budget / student_count
avg_math_score_per_school = stu_sch_df_copy.groupby("school_name")["math_score"].mean().astype("float")
avg_read_score_per_school = stu_sch_df_copy.groupby("school_name")["reading_score"].mean().astype("float")
pct_pass_math = (stu_sch_df_copy.loc[stu_sch_df_copy["math_score"] >= 70, :].groupby("school_name")["math_score"].count() / student_count) * 100
pct_pass_read = (stu_sch_df_copy.loc[stu_sch_df_copy["reading_score"] >= 70, :].groupby("school_name")["reading_score"].count() / student_count) * 100
pct_overall_pass = (stu_sch_df_copy.loc[((stu_sch_df_copy["math_score"] >= 70) & (stu_sch_df_copy["reading_score"] >= 70)), :].groupby("school_name")["student_id"].count() / student_count) * 100

summary_df = pd.DataFrame({"School Type": school_type,
                           "Total Student": student_count,
                           "Total School Budget": school_budget,
                           "Per Student Budget": per_stu_budget,
                           "Average Math Score": avg_math_score_per_school,
                           "Average Reading Score": avg_read_score_per_school,
                           "% Passing Math": pct_pass_math,
                           "% Passing Reading": pct_pass_read,
                           "% Overall Passing": pct_overall_pass})
summary_df

mapped_school_summary = summary_df.copy()
mapped_school_summary["Total School Budget"] = mapped_school_summary["Total School Budget"].map("${:,.2f}".format)
mapped_school_summary["Per Student Budget"] = mapped_school_summary["Per Student Budget"].map("${:.2f}".format)
mapped_school_summary["Average Math Score"] = mapped_school_summary["Average Math Score"].map("{:.2f}".format)
mapped_school_summary["Average Reading Score"] = mapped_school_summary["Average Reading Score"].map("{:.2f}".format)
# mapped_school_summary["% Passing Math"] = mapped_school_summary["% Passing Math"].map("{:.2f}%".format)
# mapped_school_summary["% Passing Reading"] = mapped_school_summary["% Passing Reading"].map("{:.2f}%".format)
# mapped_school_summary["% Overall Passing"] = mapped_school_summary["% Overall Passing"].map("{:.2f}%".format)
mapped_school_summary["% Passing Math"] = mapped_school_summary["% Passing Math"].map("{:.2f}".format)
mapped_school_summary["% Passing Reading"] = mapped_school_summary["% Passing Reading"].map("{:.2f}".format)
mapped_school_summary["% Overall Passing"] = mapped_school_summary["% Overall Passing"].map("{:.2f}".format)
mapped_school_summary
```

## Scores by School Spending

* Create a table that breaks down school performances based on average Spending Ranges (Per Student). Use 4 reasonable bins to group school spending. Include in the table each of the following:
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)

```python
spending_bins = [0, 585, 630, 645, 680]
spending_bins_label = ["<$585", "$585-630", "$630-645", "$645-680"]
school_summary_copy = summary_df.copy()
spending_range_per_student = pd.cut(school_summary_copy["Per Student Budget"], bins = spending_bins, labels = spending_bins_label)
school_summary_copy["Spending Ranges (Per Student)"] = spending_range_per_student
school_summary_copy

spending_range_group = school_summary_copy.groupby("Spending Ranges (Per Student)")
school_spending_summary = spending_range_group[["Average Math Score", "Average Reading Score", "% Passing Math", "% Passing Reading", "% Overall Passing"]].mean()
school_spending_summary

mapped_school_spending_summary_copy = school_spending_summary.copy()
mapped_school_spending_summary_copy["Average Math Score"] = mapped_school_spending_summary_copy["Average Math Score"].map("{:,.2f}".format)
mapped_school_spending_summary_copy["Average Reading Score"] = mapped_school_spending_summary_copy["Average Reading Score"].map("{:,.2f}".format)
mapped_school_spending_summary_copy["% Passing Math"] = mapped_school_spending_summary_copy["% Passing Math"].map("{:,.2f}".format)
mapped_school_spending_summary_copy["% Passing Reading"] = mapped_school_spending_summary_copy["% Passing Reading"].map("{:,.2f}".format)
mapped_school_spending_summary_copy["% Overall Passing"] = mapped_school_spending_summary_copy["% Overall Passing"].map("{:,.2f}".format)
mapped_school_spending_summary_copy
```

## Scores by School Size

* Perform the same operations as above, based on school size.

```python
size_bins = [0, 1000, 2000, 5000]
size_bins_label = ["Small (<1000)", "Medium (1000-2000)", "Large (2000-5000)"]
school_size_summary_copy = summary_df.copy()

school_size_binning = pd.cut(school_size_summary_copy["Total Student"], bins = size_bins, labels = size_bins_label)
school_size_summary_copy["School Size"] = school_size_binning
school_size_summary_copy

size_group = school_size_summary_copy.groupby("School Size")
size_summary = size_group[["Average Math Score", "Average Reading Score", "% Passing Math", "% Passing Reading", "% Overall Passing"]].mean()
size_summary

mapped_size_summary = size_summary.copy()
mapped_size_summary["Average Math Score"] = mapped_size_summary["Average Math Score"].map("{:,.2f}".format)
mapped_size_summary["Average Reading Score"] = mapped_size_summary["Average Reading Score"].map("{:,.2f}".format)
mapped_size_summary["% Passing Math"] = mapped_size_summary["% Passing Math"].map("{:,.2f}".format)
mapped_size_summary["% Passing Reading"] = mapped_size_summary["% Passing Reading"].map("{:,.2f}".format)
mapped_size_summary["% Overall Passing"] = mapped_size_summary["% Overall Passing"].map("{:,.2f}".format)
mapped_size_summary
```
