## Purchasing Analysis (Total)
* Run basic calculations to obtain number of unique items, average price, etc.
* Create a summary data frame to hold the results
* Optional: give the displayed data cleaner formatting

```python
unique_items = len(purchase_data["Item Name"].unique())
unique_items

avg_price = purchase_data["Price"].mean()
avg_price

num_purchase = purchase_data["Purchase ID"].count()
num_purchase

total_revenue = purchase_data["Price"].sum()
total_revenue

summary_df = pd.DataFrame({"Number of Unique Items": [unique_items],
                           "Average Price": [avg_price],
                           "Number of Purchases": [num_purchase],
                           "Total Revenue": [total_revenue]})
summary_df

mapped_summary_df = summary_df.copy()
mapped_summary_df["Average Price"] = mapped_summary_df["Average Price"].map("${:,.2f}".format)
mapped_summary_df["Total Revenue"] = mapped_summary_df["Total Revenue"].map("${:,.2f}".format)
mapped_summary_df
```

## Gender Demographics
* Percentage and Count of Male Players
* Percentage and Count of Female Players
* Percentage and Count of Other / Non-Disclosed

```python
gender_list = list(purchase_data["Gender"].unique())
# print(gender_list, type(gender_list))
sn_gender_list = purchase_data[["SN", "Gender"]]
sn_gender_list_dup = sn_gender_list.drop_duplicates(keep = "first")
sn_gender_list_dup

series_list = []

for genders in gender_list:
    gender_count = len(sn_gender_list_dup.loc[(sn_gender_list_dup["Gender"] == genders), :])
    total_player_count = len(sn_gender_list_dup["SN"])
    percent_player = (gender_count / total_player_count) * 100
    # dict_A = {"Gender": [genders], 
    #           "Total Count": [gender_count], 
    #           "Percentage of Players": [percent_player]}
    series_list.append([genders, gender_count, percent_player])
    # print(f"{genders} = {gender_count}")
    # print(f"Percent of players = {percent_player}")

df_a = pd.DataFrame(series_list, columns = ["Gender", "Total Count", "Percentage of Players"])
df_a.set_index("Gender")
df_a_sort = df_a.sort_values(by = ["Total Count"], axis = 0, ascending = False)
df_a_sort

df_a_sort["Percentage of Players"] = df_a_sort["Percentage of Players"].map("{:.2f}%".format)
df_a_sort
```

## Purchasing Analysis (Gender)
* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender
* Create a summary data frame to hold the results

```python
gender = purchase_data.set_index("Gender").groupby(["Gender"])
purchase_count = gender.SN.count()
avg_purchase_price = gender["Price"].mean()
total_purchase_value = gender["Price"].sum()

purchase_copy = purchase_data.copy()
purchase_sub = purchase_copy[["SN", "Gender"]]
dup_drop = purchase_sub.drop_duplicates(keep = "first")

total_without_dup = dup_drop.groupby(["Gender"]).count().squeeze()

avg_total_purchase_per_person = total_purchase_value / total_without_dup
avg_total_purchase_per_person

# print(purchase_count)
# print(avg_purchase_price)
# print(total_purchase_value)
# print(avg_total_purchase_per_person)
# print(type(purchase_count))
# print(type(avg_purchase_price))
# print(type(total_purchase_value))
# print(type(avg_total_purchase_per_person))

# purchase_count_df = purchase_count.to_frame()
# avg_purchase_price_df = avg_purchase_price.to_frame()
# total_purchase_value_df = total_purchase_value.to_frame()
# avg_total_purchase_per_person_df = avg_total_purchase_per_person.to_frame()
# print(purchase_count_df)
# print(avg_purchase_price_df)
# print(total_purchase_value_df)
# print(avg_total_purchase_per_person_df)
# print(type(purchase_count_df))
# print(type(avg_purchase_price_df))
# print(type(total_purchase_value_df))
# print(type(avg_total_purchase_per_person_df))



# df1 = purchase_count_df.merge(avg_purchase_price_df, left_index = True, right_index = True)
# df2 = df1.merge(total_purchase_value_df, left_index = True, right_index = True)
# df3 = df2.merge(avg_total_purchase_per_person_df, left_index = True, right_index = True)
# df3

# df3.rename(columns = {"SN": "Purchase Count",
#                       "Price_x": "Average Purchase Price",
#                       "Price_y": "Total Purchase Value",
#                       0: "Average Total Purchase per Person"}, inplace = True)
# df3

purchase_analysis = pd.DataFrame({"Purchase Count": purchase_count,
                                  "Average Purchase Price": avg_purchase_price,
                                  "Total Purchase Value": total_purchase_value,
                                  "Average Total Purchase per Person": avg_total_purchase_per_person})

purchase_analysis["Average Purchase Price"] = purchase_analysis["Average Purchase Price"].map("${:.2f}".format)
purchase_analysis["Total Purchase Value"] = purchase_analysis["Total Purchase Value"].map("${:.2f}".format)
purchase_analysis["Average Total Purchase per Person"] = purchase_analysis["Average Total Purchase per Person"].map("${:.2f}".format)
purchase_analysis

```

## Age Demographics
* Establish bins for ages
* Categorize the existing players using the age bins. Hint: use pd.cut()
* Calculate the numbers and percentages by age group
* Create a summary data frame to hold the results
* Optional: round the percentage column to two decimal points
* Display Age Demographics Table

```python
# Creating copy of original data and selecting specific columns to use
purchase_data_copy = purchase_data.copy()
purchase_data_copy_dup = purchase_data_copy[["SN", "Age"]].drop_duplicates(keep = "first")
purchase_data_copy_dup

# Age bins and bin labels
age_bins = [0, 9, 14, 19, 24, 29, 34, 39, 120]
age_bins_name = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]
age_demo_bins = pd.cut(purchase_data_copy_dup.Age, bins = age_bins, labels = age_bins_name)
purchase_data_copy_dup["Age Demographic"] = age_demo_bins
purchase_data_copy_dup

# Checking counts of players by age to verify counts in bins
purchase_data_copy_dup.loc[(purchase_data_copy_dup["Age"] < 10), :]
purchase_data_copy_dup.loc[((purchase_data_copy_dup["Age"] >= 10) & (purchase_data_copy_dup["Age"] <= 14)), :].count()

# Get count of all columns available in purchase_data_copy_dup
# age_demo_group = purchase_data_copy_dup.groupby("Age Demographic").count()

# Selected a specific column of data, namely "SN"
age_demo_group = purchase_data_copy_dup.groupby("Age Demographic")["SN"].count()
count_per_age = age_demo_group

# Count total number of players without duplicates and calculating percentage of players per age group
total_buyers = purchase_data_copy_dup["SN"].count()
age_demo_group_percent = ((count_per_age / total_buyers) * 100)

# Creating df for total count of players per group by age range and percentage of players per group by age range
age_demo_summary = pd.DataFrame({
    "Total Count": count_per_age,
    "Percentage of Players": age_demo_group_percent
})

# Formatting summary table
age_demo_summary["Percentage of Players"] = age_demo_summary["Percentage of Players"].map("{:.2f}%".format)
age_demo_summary
```
